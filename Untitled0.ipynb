```python
def read_image(path, img_size=224):
    """
    Reads an image from the given path, resizes it to the specified size,
    and normalizes pixel values to the range [0, 1].

    Args:
        path (str): Path to the image file.
        img_size (int): Target size to resize the image (default: 224).

    Returns:
        numpy.ndarray: The processed image array.
    """
    # Open the image
    img = Image.open(path).convert("RGB")
    # Resize the image
    img = img.resize((img_size, img_size))
    # Convert to a NumPy array
    img = np.array(img, dtype=np.float32)
    # Normalize pixel values
    img = img / 255.0

    return img
```

```python
def text_preprocessing(data):
    data['caption'] = data['caption'].apply(lambda x: x.lower())
    data['caption'] = data['caption'].apply(lambda x: x.replace("[^A-Za-z]",""))
    data['caption'] = data['caption'].apply(lambda x: x.replace("\s+"," "))
    data['caption'] = data['caption'].apply(lambda x: " ".join([word for word in x.split() if len(word)>1]))
    return data
```

```python
class ImageCaptioningDataset(Dataset):
    def __init__(self, dataset, processor):
        self.dataset = dataset
        self.processor = processor

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        item = self.dataset[idx]
        # Ensure item['image'] is a string, not a list
        image_path = f"/content/drive/MyDrive/image_to_text/Dataset/Images/{item['image']}"
        image = Image.open(image_path).convert("RGB")

        # Process the image and caption individually
        encoding = self.processor(images=image, text=item["caption"], padding="max_length", return_tensors="pt")
        # remove batch dimension
        encoding = {k:v.squeeze() for k,v in encoding.items()}
        return encoding

    def __getitems__(self, keys):
        """
        Custom __getitems__ to handle batching correctly.
        """
        return [self[idx] for idx in keys] # Call __getitem__ for each index
```

```python
config = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    bias="none",
    target_modules=["query","value"],
)

peft_model = get_peft_model(model, config)
```

```python
def print_trainable_parameters(model):
    """
    Prints the number of trainable parameters in the model.
    """
    trainable_params = 0
    all_param = 0
    for _, param in model.named_parameters():
        all_param += param.numel()
        if param.requires_grad:
            trainable_params += param.numel()
    print(
        f"trainable params: {trainable_params} || all params: {all_param} || trainable%: {100 * trainable_params / all_param}"
    )
```

```python
print_trainable_parameters(peft_model)
```

```python
train_dataset = ImageCaptioningDataset(dataset, processor)
train_dataloader = DataLoader(train_dataset, shuffle=True, batch_size=4)
```

```python
import os
import torch
from torch.cuda.amp import GradScaler, autocast

checkpoint_dir = "checkpoints"
os.makedirs(checkpoint_dir, exist_ok=True)

# Reduced batch size
train_dataloader = DataLoader(train_dataset, shuffle=True, batch_size=1)

# Gradient accumulation steps
gradient_accumulation_steps = 2

# Ensure model is in FP32 before creating optimizer and moving to GPU
#peft_model.to(device) # Removed this line from here
peft_model.train() # Make sure it's in train mode
peft_model.float() # Ensure model parameters are in FP32

# Optimizer initialization (after ensuring FP32 and before moving to GPU)
optimizer = torch.optim.AdamW(peft_model.parameters(), lr=5e-5)

# Now move the model to the GPU
peft_model.to(device)

# Mixed precision scaler
scaler = GradScaler()


for epoch in range(5):
    print(f"Epoch: {epoch}")
    for idx, batch in enumerate(train_dataloader):
        input_ids = batch.pop("input_ids").to(device)
        pixel_values = batch.pop("pixel_values").to(device)

        # Use autocast for mixed precision
        with autocast(dtype=torch.float16):
            outputs = peft_model(
                input_ids=input_ids,
                pixel_values=pixel_values,
                labels=input_ids,
            )

            # Normalize loss for gradient accumulation
            loss = outputs.loss / gradient_accumulation_steps

        # Scale the loss for mixed precision
        scaler.scale(loss).backward()

        if (idx + 1) % gradient_accumulation_steps == 0:
            # Step optimizer with scaled gradients
            scaler.step(optimizer)
            scaler.update()

            # Zero gradients
            optimizer.zero_grad()

        print(f"Step {idx}, Loss: {loss.item()}")

    # Save checkpoint after each epoch
    torch.save(peft_model.state_dict(), os.path.join(checkpoint_dir, f"model_epoch_{epoch}.pt"))
    print(f"Checkpoint saved for epoch {epoch}")
```
